```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load Dataset
df = sns.load_dataset('iris')

# 2. Numerical Distribution (Hint #1)
# Visualizing the distribution of sepal length
plt.figure(figsize=(8, 5))
sns.histplot(df['sepal_length'], kde=True, color='skyblue')
plt.title('Distribution of Sepal Length')
plt.show()

# 3. Categorical Analysis (Hint #2)
# Count plot for species distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='species', data=df, palette='viridis')
plt.title('Count of Each Species')
plt.show()

# 4. Identify Outliers (Hint #3)
# Box plots help identify outliers in numerical features
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title('Box Plot for Outlier Detection')
plt.show()

# 5. Feature Relationships (Hint #4)
# Correlation heatmap to understand how features relate
plt.figure(figsize=(8, 6))
correlation = df.select_dtypes(include=['number']).corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Heatmap')
plt.show()

# Summarize findings (Hint #7)
print("EDA Summary:")
print(f"- Dataset contains {df.shape[0]} samples across 3 species.")
print("- Strong positive correlation found between petal length and petal width.")
```

Data Distribution: Most features follow a normal distribution, with Sepal Length clustering between 5.0 and 6.5 units.
